## 3. Operaciones aritméticas

Python tiene los operadores: `+`, `-`, `*`, `/` (división real), `//` (división entera), `%` (módulo), `**` (potencia).  
El módulo `math` ofrece funciones como `sqrt`, `ceil`, `floor`.

````python
# Ejercicio 3a
# Dados a=23 y b=7, calcula e imprime:
#   - La división entera de a entre b
#   - El residuo (módulo) de a entre b
#   - a elevado a la potencia b (solo los primeros dígitos, usa f-string)

a, b = 23, 7

# TU CÓDIGO AQUÍ
div_int = a // b
residuo = a % b
potencia = a ** b
print(f"División entera: {div_int}")
print(f"Residuo: {residuo}")
print(f"Potencia (primeros dígitos): {potencia:.2e}")

````

````python
# Ejercicio 3b
# Usa el módulo math para calcular:
#   1. La raíz cuadrada de 144
#   2. El techo (ceil) de 7.3
#   3. El piso (floor) de 7.9
#   4. El valor de pi redondeado a 4 decimales (usa :.4f en el f-string)

import math

# TU CÓDIGO AQUÍ
raiz_cuadrada = math.sqrt(144)
techo = math.ceil(7.3)
piso = math.floor(7.9)
pi_redondeado = f"{math.pi:.4f}"

print(f"Raíz cuadrada de 144: {raiz_cuadrada:.0f}")
print(f"Techo de 7.3: {techo}")
print(f"Piso de 7.9: {piso}")
print(f"Valor de pi redondeado a 4 decimales: {pi_redondeado}")
````

## 6. Diccionarios

Los diccionarios almacenan pares **clave → valor**. Se crean con `{}`.  
Para acceder de forma segura usa `.get(clave, valor_por_defecto)`.  
Para iterar sobre pares: `for clave, valor in d.items():`

````python
# Ejercicio 6a
# Crea un diccionario que represente un curso con las claves:
#   "nombre", "codigo", "creditos", "activo"
# Luego:
#   1. Agrega la clave "estudiantes" con el valor 35
#   2. Imprime el valor de "creditos" usando .get()
#   3. Intenta obtener la clave "salon" con valor por defecto "Sin asignar"

# TU CÓDIGO AQUÍ

curso = {
    "nombre": "Álgebra Superior",
    "codigo": "M-11",
    "créditos": 12,
    "activo": True,
}

curso["Estudiantes"] = 35

print(f"Curso: {curso["nombre"]}")
print(f"Créditos: {curso.get("créditos")}")
print(f"Salón: {curso.get("salón", "Sin asignar")}")
print(f"Estudiantes inscritos: {curso["Estudiantes"]}")
````

````python
# Ejercicio 6b
# Itera sobre el diccionario del ejercicio anterior e imprime
# cada clave y su valor en el formato:  clave     : valor
# (usa f-string con alineación, ej: f"{clave:12} : {valor}")

# TU CÓDIGO AQUÍ
curso = {
    "nombre": "Álgebra Superior",
    "codigo": "M-11",
    "créditos": 12,
    "activo": True,
}

curso["Estudiantes"] = 35

for clave, valor in curso.items():
    print(f" {clave:12} : {valor}")
````

## 7. Condicionales

Estructura: `if condición:` / `elif condición:` / `else:`  
Operador ternario (en una línea): `valor_si_true if condición else valor_si_false`

````python
# Ejercicio 7a
# Escribe una función llamada clasificar_imc(imc) que reciba el
# Índice de Masa Corporal y retorne una cadena según esta escala:
#   < 18.5  → "Bajo peso"
#   18.5–24.9 → "Normal"
#   25–29.9 → "Sobrepeso"
#   >= 30   → "Obesidad"
#
# Pruébala con los valores: 16.0, 22.5, 27.3, 31.8

# TU CÓDIGO AQUÍ
def clasificar_imc(imc):
    if imc < 18.5:
        return "Bajo peso"
    elif 18.5 <= imc <= 24.9:
        return "Normal"
    elif 25 <= imc <= 29.9:
        return "Sobrepeso"
    else:
        return "Obesidad"
    
valores_imc = [16.0, 22.5, 27.3, 31.8]

for valor in valores_imc:
    clasificacion = clasificar_imc(valor)
    print(f"IMC: {valor} -> Clasificación: {clasificacion}")
````

````python
# Ejercicio 7b
# Usa el operador ternario para determinar si un número es par o impar.
# Pruébalo con los números 4, 7 y 0.
# Salida esperada:  4 es par  /  7 es impar  /  0 es par

# TU CÓDIGO AQUÍ

numeros = [4, 7, 0]

resultados = [f"{num} es {'par' if num % 2 == 0 else 'impar'}" for num in numeros]

print("  /  ".join(resultados))

#Si el / se refería a un salto de línea, entonces:

print(f"\n".join(resultados))

````

## 8. Ciclos

- `for elemento in iterable:` — recorre cada elemento
- `for i in range(inicio, fin, paso):` — secuencia numérica
- `for i, val in enumerate(lista):` — índice + valor
- `while condición:` — repite mientras la condición sea verdadera
- **List comprehension:** `[expresión for x in iterable if condición]`

````python
# Ejercicio 8a
# Usa un for con enumerate() para imprimir la siguiente lista de planetas
# numerados desde 1.
# Salida esperada:
#   [1] Mercurio
#   [2] Venus
#   ...

planetas = ["Mercurio", "Venus", "Tierra", "Marte", "Júpiter", "Saturno", "Urano", "Neptuno"]

# TU CÓDIGO AQUÍ

for i, planeta in enumerate(planetas, start=1):
    print(f"[{i}] {planeta}")
````

````python
# Ejercicio 8b
# Usa un ciclo while para imprimir la tabla de multiplicar del 7
# (del 7x1 hasta el 7x10).
# Formato:  7 x 1 = 7

# TU CÓDIGO AQUÍ
n = 1
while n <= 10:
    producto = 7 * n
    print(f"7 x {n} = {producto}")
    n += 1
````

````python
# Ejercicio 8c — List comprehension
# En una sola línea, crea las siguientes listas y luego imprímelas:
#   1. Los cubos de los números del 1 al 8  (x³)
#   2. Los números del 1 al 30 que sean divisibles por 3 pero NO por 9

# TU CÓDIGO AQUÍ
cubos = [x**3 for x in range(1, 9)]
divisibles = [x for x in range(1, 31) if x % 3 == 0 and x % 9 != 0]

print(f"Cubos del 1 al 8: {cubos}")
print(f"Números del 1 al 30 divisibles por 3 pero no por 9: {divisibles}")
````

Una función puede retornar **múltiples valores** (como tupla) y llamarse a sí misma (**recursión**).

````python
# Ejercicio 9a
# Escribe una función convertir_temperatura(grados, escala="C") que convierta:
#   - Si escala="C"  → convierte Celsius a Fahrenheit: F = C * 9/5 + 32
#   - Si escala="F"  → convierte Fahrenheit a Celsius: C = (F - 32) * 5/9
#
# Pruébala con: 100°C, 0°C, 32°F y 212°F

# TU CÓDIGO AQUÍ
def convertir_grados(grados):
    escala = grados[-1].upper()
    grados = grados.removesuffix("°C").removesuffix("°F")
    grados = float(grados)

    if escala == "C":
        return str(grados * 9/5 + 32) + " °F"
    elif escala == "F":
        return str((grados - 32) * 5/9) + " °C"
    else:
        return "Escala no reconocida" 
    
print(convertir_grados("100°C"))
print(convertir_grados("0°C"))
print(convertir_grados("32°F"))
print(convertir_grados("212°F"))

````

````python
# Ejercicio 9b
# Escribe una función resumen_lista(lista) que retorne
# tres valores: (minimo, maximo, promedio).
# Úsala con: [12, 45, 7, 89, 34, 56, 23]
# Imprime los tres resultados usando desempaquetado.

# TU CÓDIGO AQUÍ
def resumen_lista(lista):
    return min(lista), max(lista), sum(lista) / len(lista)

numeros = [12, 45, 7, 89, 34, 56, 23]

minimo, maximo, promedio = resumen_lista(numeros)

print(f"Mínimo: {minimo}, Máximo: {maximo}, Promedio: {promedio:.2f}")
````

````python
# Ejercicio 9c — Recursión
# Escribe una función recursiva fibonacci(n) que retorne
# el n-ésimo número de la sucesión de Fibonacci.
# Casos base: fibonacci(0)=0, fibonacci(1)=1
# Caso recursivo: fibonacci(n) = fibonacci(n-1) + fibonacci(n-2)
#
# Imprime los primeros 10 números: fibonacci(0) ... fibonacci(9)

# TU CÓDIGO AQUÍ

def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
for i in range(10):
    print(fibonacci(i))
````

## 10. Manejo de errores

```python
try:
    # código que puede fallar
except TipoDeError as e:
    # qué hacer si ocurre el error
finally:
    # se ejecuta siempre
```
Para lanzar un error explícitamente: `raise ValueError("mensaje")`

````python
# Ejercicio 10a
# Escribe una función convertir_entero(valor) que intente convertir
# el parámetro a entero usando int().
#   - Si tiene éxito, retorna el entero.
#   - Si ocurre un ValueError, imprime un mensaje de error y retorna None.
#   - El bloque finally debe imprimir "[intento con: <valor>]" siempre.
#
# Pruébala con: "42", "3.14", "hola", 100

# TU CÓDIGO AQUÍ
def convertir_entero(valor):
    try:
        return int(float(valor))    # Ajuste para que convertir números con decimales también funcione
    except ValueError:
        print(f"Error: No se pudo convertir '{valor}' a entero.")
    finally:
        print(f"[intento con: {valor}]")

print(convertir_entero("42"))
print()
print(convertir_entero("3.14"))
print()
print(convertir_entero("hola"))
print()
print(convertir_entero(100))
````

````python
# Ejercicio 10b
# Escribe una función validar_edad(edad) que:
#   - Lance un TypeError  si edad no es int
#   - Lance un ValueError si edad < 0 o edad > 150
#   - Retorne True si la edad es válida
#
# Pruébala dentro de un try/except con: 25, -5, 200, "veinte"

# TU CÓDIGO AQUÍ

def validar_edad(edad):
    if not isinstance(edad, int):
        raise TypeError("La edad debe ser un número entero.")
    if edad < 0 or edad > 150:
        raise ValueError("La edad debe estar entre 0 y 150.")
    return True

prueba_edades = [25, -5, 200, "veinte"]

for edad in prueba_edades:
    try:
        if validar_edad(edad):
            print(f"Edad {edad} es válida.")
    except (TypeError, ValueError) as e:
        print(f"Error al validar edad '{edad}': {e}")
````

Combina lo aprendido en un mini-programa completo.

````python
# Ejercicio integrador
#
# Crea una función analizar_curso(estudiantes) que reciba una lista de
# diccionarios. Cada diccionario representa un estudiante con las claves:
#   "nombre" (str), "notas" (lista de ints)
#
# La función debe:
#   1. Calcular el promedio de cada estudiante (retorna None si notas está vacía)
#   2. Clasificar al estudiante: promedio >= 61 → "Aprobado", si no → "Reprobado"
#   3. Imprimir una tabla con: nombre, promedio (2 decimales) y estado
#   4. Al final, imprimir cuántos aprobaron y cuántos reprobaron
#
# Usa el siguiente listado de prueba:

curso = [
    {"nombre": "Ana López",    "notas": [85, 90, 78, 92]},
    {"nombre": "Carlos Ruiz",  "notas": [55, 48, 60, 52]},
    {"nombre": "María Paz",    "notas": [70, 75, 68, 80]},
    {"nombre": "Luis Gómez",   "notas": [95, 98, 100, 92]},
    {"nombre": "Sara Díaz",    "notas": [40, 55, 50, 45]},
    {"nombre": "Pedro Alva",   "notas": []},
]

# TU CÓDIGO AQUÍ

def analizar_curso(estudiantes):
    aprobados = 0
    reprobados = 0

    for estudiante in estudiantes:

        notas = estudiante.get("notas")

        if not notas:
            promedio = None
        else:
            promedio = sum(notas) / len(notas)

        if promedio is None:
            estado = "Sin notas registradas"
        elif promedio >= 61:
            estado = "Aprobado"
            aprobados += 1
        else: 
            estado = "Reprobado"
            reprobados += 1
        
        print(f"Nombre: {estudiante.get("nombre"):15} | Promedio: {promedio if promedio is not None else "N/A":<8} | Estado: {estado}")

    print("\nResumen del curso:")
    print(f"Aprobados: {aprobados}")
    print(f"Reprobados: {reprobados}")

analizar_curso(curso)


````
